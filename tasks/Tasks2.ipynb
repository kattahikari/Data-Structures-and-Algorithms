### Задание №1
Массив A состоит из целых положительных чисел длин отрезков. Составьте из трех отрезков такой треугольник, чтобы его периметр был максимально возможным. Если невозможно составить треугольник с положительной площадью функция возвращает 0.

```python
import numpy as np

user_min = int(input())
user_max = int(input())
n = int(input())

A = np.random.randint(user_min, user_max, size=(n, 3)) 
max_num_index = np.argmax(A, axis=1)
 
max_per = 0 
max_array_index = 0 
 
print('Исходные данные:\n' + str(A))
 
for i in range(len(A)):
    if max_num_index[i] != 0:
        A[i,0], A[i, max_num_index[i]] = A[i, max_num_index[i]], A[i,0]
        
for i in range(len(A)):
    if (A[i,0]) < (A[i,1] + A[i,2]):
        if (A[i,0] + A[i,1] + A[i,2]) > max_per: 
            max_per, max_array_index = A[i,0] + A[i,1] + A[i,2], i
            
if max_per == 0:
    print('Такого треугольника не существует')
else:
    print('У треугольника со сторонами', A[max_array_index,0], A[max_array_index,1], A[max_array_index,2], 'максимальный периметр', max_per)
```

### Задание №2
Дан массив неотрицательных целых чисел nums. Расположите их в таком порядке, чтобы вместе они образовали максимально возможное число. Замечание: Результат может быть очень большим числом, поэтому представьте его как string, а не integer.

```python
import random

#Генерация массива 

min_limit = int(input())
max_limit = int(input())
m = int(input())

nums = np.random.randint(min_limit,max_limit,(m))

print('Исходный массив: \n' + str(nums))

nums = [str(i) for i in nums]

nums_to_fix = []

for i in nums:
    i = list(''.join(i))
    for j in i:
        nums_to_fix.append(j)

nums = sorted([int(i) for i in nums_to_fix],reverse = True)

string = ""

for i in nums:
    string += str(i)

print('Получившееся число: \n' + string)
```

### Задание №3
Дана матрица mat размером m * n, значения целочисленные. Напишите функцию, сортирующую каждую диагональ матрицы по возрастанию и возвращающую получившуюся матрицу.

```python
import numpy as np
import random

#Генерация матрицы 

min_limit = int(input())
max_limit = int(input())
m = int(input())
n = int(input())

mat = np.random.randint(min_limit,max_limit,(m,n))
print('Исходная матрица: \n' + str(mat))



#Сортировка главной диагонали

for i in range(m):
    for j in range(i+1,n):
        if mat[i][i] > mat[j][j]:
            mat[i][i],mat[j][j] = mat[j][j],mat[i][i]

#Сортировка побочной диагонали

for i in range(m):
    for j in range(i+1,n):
        if mat[i][-i-1] < mat[j][-j-1]:
            mat[i][-i-1],mat[j][-j-1] = mat[j][-j-1],mat[i][-i-1]
            
print('Матрица с отсортированными диагоналями: \n' + str(mat))       
    
```
