–ö–æ—Ä–æ–≤–∞ –ë–µ—Å—Å–∏ —Ç–æ–ª—å–∫–æ —á—Ç–æ –ø–µ—Ä–µ—Ö–≤–∞—Ç–∏–ª–∞ —Å–æ–æ–±—â–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –§–µ—Ä–º–µ—Ä –î–∂–æ–Ω –æ—Ç–ø—Ä–∞–≤–∏–ª –ë—É—Ä–≥–µ—Ä –ö—É–∏–Ω! –ë–µ—Å—Å–∏ —É–≤–µ—Ä–µ–Ω–∞, —á—Ç–æ –≤ –Ω–µ–º —Å–∫—Ä—ã—Ç–æ —Å–µ–∫—Ä–µ—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ.

–ù–∞–ø—Ä–∏–º–µ—Ä, –≤ —Å—Ç—Ä–æ–∫–µ aaabb, a —Å–∫—Ä—ã—Ç–∞ 3 —Ä–∞–∑–∞, b —Å–∫—Ä—ã—Ç–∞ 2 —Ä–∞–∑–∞, ab —Å–∫—Ä—ã—Ç–∞ 6 —Ä–∞–∑, aa —Å–∫—Ä—ã—Ç–∞ 3 —Ä–∞–∑–∞, bb —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑, aab —Å–∫—Ä—ã—Ç–∞ 2 —Ä–∞–∑–∞, aaa —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑, abb —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑, aaab —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑, aabb —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑ –∏ aaabb —Å–∫—Ä—ã—Ç–∞ 1 —Ä–∞–∑. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Å–∫—Ä—ã—Ç–æ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏—è —Ä–∞–≤–Ω–æ 6.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ùë†, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ (1‚â§|ùë†|‚â§105) ‚Äî —Ç–µ–∫—Å—Ç, –∫–æ—Ç–æ—Ä—ã–π –ø–µ—Ä–µ—Ö–≤–∞—Ç–∏–ª–∞ –ë–µ—Å—Å–∏.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Å–µ–∫—Ä–µ—Ç–Ω–æ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏—è.

```python
from collections import defaultdict


def task1(in1):
    d = defaultdict(int)
    for c in in1:
        for a in range(97,123):
            a = chr(a)
            d[a + c] += d[a]
        d[c] += 1
    return str(max(d.values()))

print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "aaabb")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task1("aaabb"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "usaco")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task1("usaco"))

```

**–ó–∞–¥–∞–Ω–∏–µ 2**

–í–∞–º –∑–∞–¥–∞–Ω—ã –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ ùë† –∏ ùë°, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ —Å—Ç—Ä–æ—á–Ω—ã—Ö –±—É–∫–≤ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞. –í—ã —Å–æ–±–∏—Ä–∞–µ—Ç–µ—Å—å –ø–æ—Å–∏–º–≤–æ–ª—å–Ω–æ –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å —Å—Ç—Ä–æ–∫—É ùë†, –Ω–∞—á–∏–Ω–∞—è —Å –ø–µ—Ä–≤–æ–≥–æ —Å–∏–º–≤–æ–ª–∞ –∏ –∑–∞–∫–∞–Ω—á–∏–≤–∞—è –ø–æ—Å–ª–µ–¥–Ω–∏–º.

–ö–æ–≥–¥–∞ –≤—ã —Å–æ–±–∏—Ä–∞–µ—Ç–µ—Å—å –Ω–∞–ø–µ—á–∞—Ç–∞—Ç—å –∫–∞–∫–æ–π-—Ç–æ —Å–∏–º–≤–æ–ª, –≤–º–µ—Å—Ç–æ —Ç–æ–≥–æ, —á—Ç–æ–±—ã –Ω–∞–∂–∞—Ç—å –Ω–∞ –∫–Ω–æ–ø–∫—É, –ø–µ—á–∞—Ç–∞—é—â—É—é —ç—Ç–æ—Ç —Å–∏–º–≤–æ–ª, –≤—ã –º–æ–∂–µ—Ç–µ –Ω–∞–∂–∞—Ç—å –∫–Ω–æ–ø–∫—É ¬´Backspace¬ª. –ù–∞–∂–∞—Ç–∏–µ –Ω–∞ —ç—Ç—É –∫–Ω–æ–ø–∫—É —É–¥–∞–ª—è–µ—Ç –ø–æ—Å–ª–µ–¥–Ω–∏–π –Ω–∞–ø–µ—á–∞—Ç–∞–Ω–Ω—ã–π —Å–∏–º–≤–æ–ª, –∫–æ—Ç–æ—Ä—ã–π –µ—â–µ –Ω–µ –±—ã–ª —É–¥–∞–ª–µ–Ω (–∏–ª–∏ –Ω–∏—á–µ–≥–æ –Ω–µ –¥–µ–ª–∞–µ—Ç, –µ—Å–ª–∏ –≤—Å–µ –Ω–∞–ø–µ—á–∞—Ç–∞–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã —É–∂–µ —É–¥–∞–ª–µ–Ω—ã –∏–ª–∏ –≤—ã –µ—â–µ –Ω–µ –Ω–∞–ø–µ—á–∞—Ç–∞–ª–∏ –Ω–∏ –æ–¥–Ω–æ–≥–æ —Å–∏–º–≤–æ–ª–∞). –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ —Å—Ç—Ä–æ–∫–∞ ùë† ‚Äî ¬´abcbd¬ª, –∏ –≤—ã –Ω–∞–∂–∏–º–∞–µ—Ç–µ –Ω–∞ –∫–Ω–æ–ø–∫—É Backspace –≤–º–µ—Å—Ç–æ –ø–µ—á–∞—Ç–∏ –ø–µ—Ä–≤–æ–≥–æ –∏ —á–µ—Ç–≤–µ—Ä—Ç–æ–≥–æ —Å–∏–º–≤–æ–ª–∞, –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø–æ–ª—É—á–∏—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ ¬´bd¬ª (–ø–µ—Ä–≤–æ–µ –Ω–∞–∂–∞—Ç–∏–µ Backspace –Ω–µ —É–¥–∞–ª–∏—Ç –Ω–∏ –æ–¥–Ω–æ–≥–æ —Å–∏–º–≤–æ–ª–∞, –∞ –≤—Ç–æ—Ä–æ–µ –Ω–∞–∂–∞—Ç–∏–µ —ç—Ç–æ–π –∫–Ω–æ–ø–∫–∏ —É–¥–∞–ª–∏—Ç —Å–∏–º–≤–æ–ª ¬´c¬ª). –î—Ä—É–≥–æ–π –ø—Ä–∏–º–µ—Ä: –µ—Å–ª–∏ ùë† —Ä–∞–≤–Ω–∞ ¬´abcaa¬ª, –∏ –≤—ã –Ω–∞–∂–∏–º–∞–µ—Ç–µ Backspace –≤–º–µ—Å—Ç–æ –¥–≤—É—Ö –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –±—É–∫–≤, –ø–æ–ª—É—á–∞–µ—Ç—Å—è ¬´a¬ª.

–í—ã –¥–æ–ª–∂–Ω—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –º–æ–∂–Ω–æ –ª–∏ –ø–æ–ª—É—á–∏—Ç—å —Å—Ç—Ä–æ–∫—É ùë°, –µ—Å–ª–∏ –≤—ã –ø–æ–ø—Ä–æ–±—É–µ—Ç–µ –Ω–∞–±—Ä–∞—Ç—å —Å—Ç—Ä–æ–∫—É ùë†, –Ω–∞–∂–∏–º–∞—è Backspace –≤–º–µ—Å—Ç–æ –Ω–∞–∂–∞—Ç–∏—è –∫–Ω–æ–ø–æ–∫, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –Ω–µ–∫–æ—Ç–æ—Ä—ã–º (–≤–æ–∑–º–æ–∂–Ω–æ, –Ω–∏ –æ–¥–Ω–æ–º—É) –±—É–∫–≤–∞–º —Å—Ç—Ä–æ–∫–∏ ùë†.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–¥–∞–Ω–æ –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëû (1‚â§ùëû‚â§105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç —Å—Ç—Ä–æ–∫—É ùë† (1‚â§|ùë†|‚â§105). –ö–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª —Å—Ç—Ä–æ–∫–∏ ùë† ‚Äî —Å—Ç—Ä–æ—á–Ω–∞—è –±—É–∫–≤–∞ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞.

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç —Å—Ç—Ä–æ–∫—É ùë° (1‚â§|ùë°|‚â§105). –ö–∞–∂–¥—ã–π —Å–∏–º–≤–æ–ª —Å—Ç—Ä–æ–∫–∏ ùë° ‚Äî —Å—Ç—Ä–æ—á–Ω–∞—è –±—É–∫–≤–∞ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞.

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—É–º–º–∞—Ä–Ω–æ–µ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–∏–º–≤–æ–ª–æ–≤ –≤–æ –≤—Å–µ—Ö —Å—Ç—Ä–æ–∫–∞—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2‚ãÖ105.

```python
def task2(in1, in2):   
    l,r=len(in1)-1,len(in2)-1
    while l>=0 and r>=0:
        if in1[l]==in2[r]:
            l-=1
            r-=1
        else:
            l-=2
 
    if r == -1:
        result = "YES"
    else:
        result = "NO"
    return result

print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "ababa" + '\n' + "ba")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task2("ababa", "ba"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "ababa" + '\n' + "bb")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task2("ababa", "bb"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "aaa" + '\n' + "aaaa")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task2("aaa", "aaaa"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "aababa" + '\n' + "ababa")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task2("aababa", "ababa"))
```

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—É–º–º–∞ ùëõ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2‚ãÖ105.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–î–ª—è –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤—ã–≤–µ–¥–∏—Ç–µ –¥–≤–µ —Å—Ç—Ä–æ–∫–∏:

–≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ‚Äî –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É, –∫–æ—Ç–æ—Ä–∞—è –º–æ–≥–ª–∞ –±—ã—Ç—å –ø–æ–¥–∞—Ä–µ–Ω–∞ –ü–æ–ª–∏–∫–∞—Ä–ø—É –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ; –≤–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ‚Äî –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É, –∫–æ—Ç–æ—Ä–∞—è –º–æ–≥–ª–∞ –±—ã—Ç—å –ø–æ–¥–∞—Ä–µ–Ω–∞ –ü–æ–ª–∏–∫–∞—Ä–ø—É –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ.

```python
from collections import deque

def task5(in1, in2):
    in1 = int(in1)
    in2 = in2.split(' ')
    in2 = list(map(int, in2))
    minn, maxx, a, p1, p2 = [ ],  [ ], 0, deque(), deque()
    for i in range(in1):
        c = in2[i]
        if c == a: minn.append(p1.popleft()), maxx.append(p2.pop())
        else:
            minn.append(c), maxx.append(c)
            for j in range(a + 1, c):
                p1.append(j)
                p2.append(j)
            a = c

    return ' '.join(map(str, minn)) + '\n' + ' '.join(map(str, maxx))

print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "7" + '\n' + "3 3 4 4 7 7 7")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task5("7", "3 3 4 4 7 7 7"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "4" + '\n' + "1 2 3 4")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task5("4", "1 2 3 4"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "7" + '\n' + "3 4 5 5 5 7 7")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task5("7", "3 4 5 5 5 7 7"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "1" + '\n' + "1")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task5("1", "1"))  
```

```python

n = int(input())
words = [input() for i in range(n)]
ans = 0
for i in "abcde":
    counts = sorted([-len(word) + 2 * word.count(i) for word in words])[::-1]
    su, n = 0, 0   
    for count in counts:
        su += count
        n += 1
        if su <= 0:
            n -= 1
            break 
    ans = max(ans, n)
    
print(ans)
```

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—É–º–º–∞ ùëõ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 3‚ãÖ105.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–î–ª—è –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤—ã–≤–µ–¥–∏—Ç–µ ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –≥–¥–µ ùëñ-–µ —á–∏—Å–ª–æ —è–≤–ª—è–µ—Ç—Å—è ùëñ-—É–¥–∏–≤–∏—Ç–µ–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º –º–∞—Å—Å–∏–≤–∞.

```python
def task9(in1, in2):
    in1 = int(in1)
    in2 = in2.split(' ')
    in2 = list(map(int, in2))
    check=[0]*(in1+1)
    last=[0]*(in1+1)
    ans=[-1]*(in1+1)
    for i in range(in1):
        check[in2[i]]=max(check[in2[i]],i-last[in2[i]]+1)
        last[in2[i]]=i+1
    for i in range(1,in1+1):
        check[i]=max(check[i],in1-last[i]+1)
        j=check[i]
        while j<=in1 and ans[j]==-1:
            ans[j]=i
            j+=1
    return " ".join(map(str, ans))

print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "5" + '\n' + "1 2 3 4 5")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task9("5", "1 2 3 4 5"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "5" + '\n' + "4 4 4 4 2")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task9("5", "4 4 4 4 2"))
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: " + '\n' + "6" + '\n' + "1 3 1 5 3 1")
print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:" + '\n' + task9("6", "1 3 1 5 3 1"))
```

–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ n, m (2‚Äâ‚â§‚Äân‚Äâ‚â§‚Äâ3¬∑105; 1‚Äâ‚â§‚Äâm‚Äâ‚â§‚Äâ3¬∑105) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—ã—Ü–∞—Ä–µ–π –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–∞–∂–µ–Ω–∏–π. –í –∫–∞–∂–¥–æ–π –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö m —Å—Ç—Ä–æ–∫ –∑–∞–ø–∏—Å–∞–Ω–æ —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ li,‚Äâri,‚Äâxi (1‚Äâ‚â§‚Äâli‚Äâ<‚Äâri‚Äâ‚â§‚Äân; li‚Äâ‚â§‚Äâxi‚Äâ‚â§‚Äâri) ‚Äî –æ–ø–∏—Å–∞–Ω–∏–µ i-–≥–æ —Å—Ä–∞–∂–µ–Ω–∏—è.

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏.\ –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –≤ –∫–∞–∂–¥–æ–º —Å—Ä–∞–∂–µ–Ω–∏–∏ —É—á–∞—Å—Ç–≤–æ–≤–∞–ª–∏ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–∞ —Ä—ã—Ü–∞—Ä—è.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ n —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. –ï—Å–ª–∏ i-—ã–π —Ä—ã—Ü–∞—Ä—å –±—ã–ª –ø–æ–±–µ–∂–¥–µ–Ω, —Ç–æ i-–æ–µ —á–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–∞–≤–Ω–æ –Ω–æ–º–µ—Ä—É —Ä—ã—Ü–∞—Ä—è, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–±–µ–¥–∏–ª —Ä—ã—Ü–∞—Ä—è —Å –Ω–æ–º–µ—Ä–æ–º i. –ï—Å–ª–∏ i-—ã–π —Ä—ã—Ü–∞—Ä—å —è–≤–ª—è–µ—Ç—Å—è –ø–æ–±–µ–¥–∏—Ç–µ–ª–µ–º —Ç—É—Ä–Ω–∏—Ä–∞, —Ç–æ i-–æ–µ —á–∏—Å–ª–æ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–∞–≤–Ω–æ 0.

```python
knights, battles = map(int, input().split())
res = [i + 1 for i in range(knights)]
ans = [0] * knights
for i in range(battles):
    l, r, x = map(int, input().split())
    li ,xi = l - 1, x - 1
    while li < r:
        if ans[li] == 0 and li != xi:
            ans[li] = x
        a = res[li]
        if li < xi:
            res[li] = xi
        else:
            res[li] = r
        li = a
print(*ans)
```
