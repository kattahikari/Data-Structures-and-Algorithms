```python
#Бинарный поиск

import numpy as np
import random

user_min_limit = int(input())
user_max_limit = int(input())
user_n = int(input())
number_to_find = int(input())

arr = np.random.randint(user_min_limit,user_max_limit,(user_n))

if number_to_find in arr:
    pass
else:
    np.append(arr,number_to_find)

arr = sorted(arr)

def binary_search(arr, value):
    first = 0
    last = len(arr)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if arr[mid] == value:
            index = mid
        else:
            if value<arr[mid]:
                last = mid -1
            else:
                first = mid +1
    return index

print(arr)
binary_search(arr, number_to_find)
```

```python
#Бинарное дерево

import numpy as np
import random

user_min_limit = int(input())
user_max_limit = int(input())
user_n = int(input())

arr = np.random.randint(user_min_limit,user_max_limit,(user_n))

random_index = random.randint(0, len(arr) - 1)

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


def inorder(root):
    if root is not None:
        inorder(root.left)
        print(str(root.key) + "->", end=' ')
        inorder(root.right)

def insert(node, key):
    if node is None:
        return Node(key)

    if key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)
    return node

def minValueNode(node):
    current = node
    while(current.left is not None):
        current = current.left
    return current

def deleteNode(root, key):
    if root is None:
        return root

    if key < root.key:
        root.left = deleteNode(root.left, key)
    elif(key > root.key):
        root.right = deleteNode(root.right, key)
    else:     
        if root.left is None:
            temp = root.right
            root = None
            return temp

        elif root.right is None:
            temp = root.left
            root = None
            return temp
        temp = minValueNode(root.right)
        root.key = temp.key
        root.right = deleteNode(root.right, temp.key)
    return root

root = Node(random_index)
print("Исходный массив:" + str(arr))
print("Случайно выбранный корень:" + str(random_index))

for i in range(len(arr)):
    root = insert(root, i)
    
inorder(root)
```

```python
#Метод Фибоначчи

import numpy as np
import random

user_min_limit = int(input())
user_max_limit = int(input())
user_n = int(input())

arr = sorted(np.random.randint(user_min_limit,user_max_limit,(user_n)))
random_index = random.choice(arr)

def fibonacci_search(arr, random_index):
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(arr)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(arr)-1))
        if (arr[i] < random_index):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (arr[i] > random_index):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else:
            return i
    if(fibM_minus_1 and index < (len(arr)-1) and arr[index+1] == random_index):
        return index+1
    return -1

print("Исходный массив:" + str(arr))
print("Случайный элемент массива:" + str(random_index))
print(fibonacci_search(arr, random_index))
```

```python
#Интерполяционный

import numpy as np
import random

user_min_limit = int(input())
user_max_limit = int(input())
user_n = int(input())

arr = sorted(np.random.randint(user_min_limit,user_max_limit,(user_n)))
random_index = random.choice(arr)


def interpolation_search(arr, random_index):
    low = 0
    high = (len(arr) - 1)
    while low <= high and random_index >= arr[low] and random_index <= arr[high]:
        index = low + int(((float(high - low) / (arr[high] - arr[low])) * (random_index - arr[low])))
        if arr[index] == random_index:
            return index
        if arr[index] < random_index:
            low = index + 1;
        else:
            high = index - 1;
    return -1

print("Исходный массив:" + str(arr))
print("Случайный элемент массива:" + str(random_index))
print(fibonacci_search(arr, random_index))
```

```python

months = ["January", "February", "March", "May", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

def ascii_sum(month):
    sum = 0
    for el in month: sum += ord(el)
    print(sum)

for month in months:
    ascii_sum(month)
    
month_hash_list = [0 for i in range(len(months))]
def simple_hash(month, size, p = 0):
    result = 0
    for el in month:
        result += ord(el)
    return (result + p) % size

for month in months:
    i = simple_hash(month, len(months))
    p = 1
    while month_hash_list[i] != 0:
        i = simple_hash(month, len(months), p)
        p += 1
    month_hash_list[i] = month

print(month_hash_list)
for month in months:
    i = simple_hash(month, len(months))
    p = 1
    while month_hash_list[i] != month:
        i = simple_hash(month, len(months), p)
        p += 1
    print(month_hash_list[i], i)

for month in months:
    print(month, simple_hash(month, len(months)), month == month_hash_list[simple_hash(month, len(months))])
```

```python
#Рехэширование с помощью псевдослучайных чисел

month_hash_list = [0 for i in range(len(months))]

def generate_random_num(month, size, p):
    a = 2
    b = 5
    return ((p ** b) // (size ** a)) + len(month) ** (b - a)

def random_hash(month, size, p = 0):
    result = 0
    for el in month:
        result += ord(el)
    return (result + generate_random_num(month, size, p)) % size

for month in months:
    i = random_hash(month, len(months))
    p = 1
    while month_hash_list[i] != 0:
        i = random_hash(month, len(months), p)
        p += 1
    month_hash_list[i] = month

print(month_hash_list)
for month in months:
    i = random_hash(month, len(months))
    p = 1
    while month_hash_list[i] != month:
        i = random_hash(month, len(months), p)
        p += 1
    print(month_hash_list[i], i)

for month in months:
    print(month, random_hash(month, len(months)), month == month_hash_list[random_hash(month, len(months))])
```

```python
#Метод цепочек

class Node():
    
    def __init__(self, data = None):
        self.data = data
        self.next = None
    
    def add(self, data):
        if self.data is None:
            self.data = data
        elif self.next is None:
            self.next = Node(data)
        else:
            self.next.add(data)
    def print_Node(self):
        if self.data:
            print(self.data, end = ' ')
            if self.next:
                self.next.print_Node()
            else:
                return
month_hash_list = [Node() for i in range(len(months))]

def node_hash(month, size):
    result = 0
    for el in month:
        result += ord(el)
    return result % size

for month in months:
    i = node_hash(month, len(months))
    month_hash_list[i].add(month)
    
for i in range(len(month_hash_list)):
    print(i, end = ' ')
    month_hash_list[i].print_Node()
    print()
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
field = [[0 for i in range(8)] for j in range(8)]

def set_queen(i, j):
    for k in range(8):
        field[k][j] += 1
        field[i][k] += 1
        
        if 0 <= i + j - k <= 7:
            field[i + j - k][k] += 1
        if 0 <= i - j + k <= 7:
            field[i - j + k][k] += 1
    field[i][j] = -1

def delete_queen(i, j):
    for k in range(8):
        field[k][j] -= 1
        field[i][k] -= 1
        
        if 0 <= i + j - k <= 7:
            field[i + j - k][k] -= 1
        if 0 <= i - j + k <= 7:
            field[i - j + k][k] -= 1
    field[i][j] = 0
    
def print_answer():
    key = "ABCDEFGH"
    result = []
    for i in range(8):
        for j in range(8):
            if field[i][j] == -1:
                result.append(key[j] + str(i + 1))
    print(' '.join(result))

def chess_queens(i = 0):
    for j in range(8):
        if field[i][j] == 0:
            set_queen(i, j)
            if i == 7:
                print_answer()
            else:
                chess_queens(i + 1)
            delete_queen(i, j)

    
chess_queens()
```
